```python
# univariate variable

data = {'Salary':[25000,30000,37000,28000,39000,48000,55000,52000,35000,26000,27000,31000]}
```

```python
s = pd.DataFrame(data)
```

```python
s.head()
```

```python
s.shape
```

```python
# Line Plot
plt.plot(s['Salary'])
plt.show()
```

```python
plt.plot(s['Salary'], color = 'red' , marker = 'o' , linestyle = '--', linewidth = 2)
plt.grid()
plt.show()
```

```python
# Histograms

plt.hist(s['Salary'])
plt.show()
```

```python
plt.hist(s['Salary'],bins=5,color='yellow')
plt.show()
```

```python
plt.boxplot(s['Salary'])
plt.show()
```

```python
s.loc[20] = 0
```

```python
plt.boxplot(s['Salary'])
plt.show()
```

```python
s.drop(index=20, inplace=True)
```

```python
plt.boxplot(s['Salary'])
plt.show()
```
