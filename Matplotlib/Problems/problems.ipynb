```python
# 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,label = 'Problem-1')
plt.xlabel = 'x'
plt.ylabel = 'y'
plt.title = 'draw a line'
plt.show()
```

```python
# 2. Write a Python program to draw a line using given axis values with suitable label in the x axis, y axis and a title

x = [10,20,30,40,50]
y=[60,70,80,90,100]
plt.plot(x,y, label = 'Line Graph')
plt.xlabel = 'x axis'
plt.ylabel = 'y axis'
plt.title = 'Line plot'
plt.legend()
plt.show()
```

```python
x = []
y = []

with open(r"test.txt", "r") as file:
    for line in file:
        a, b = line.split()
        x.append(int(a))
        y.append(int(b))

plt.plot(x, y)
plt.xlabel = 'x'
plt.ylabel = 'y'
plt.title = 'line graphh'
plt.show()
```

```python
# 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [5, 6, 7, 8, 9]

plt.plot(x, y1, label="Line 1", color = 'orange')
plt.plot(x, y2, label="Line 2" , color = 'black')
plt.plot(x, y3, label="Line 3", color = 'red')

plt.xlabel="X Axis"
plt.ylabel="Y Axis"
plt.title="Multiple Line Plot"

plt.legend()

plt.show()
```

```python
# 6. Write a Python program to plot two or more lines with legends, different widths and colors.

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [8, 18, 28, 38, 48]
y3 = [5, 15, 25, 35, 45]

plt.plot(x, y1, color='orange', linewidth=4, label="Product A")
plt.plot(x, y2, color='white', linewidth=4, label="Product B")
plt.plot(x, y3, color='green', linewidth=4, label="Product C")



plt.xlabel="Time"
plt.ylabel="Performance"
plt.title="Mulgiple lines"

plt.legend()

plt.show()
```

```python
# 7. Write a Python program to plot two or more lines with different styles
x = [1, 2, 3, 4, 5]
y1 = [5, 10, 15, 20, 25]
y2 = [5, 8, 12, 18, 22]
y3 = [3, 6, 9, 12, 15]

plt.plot(x, y1, linestyle='-', linewidth=2, label="Solid Line")
plt.plot(x, y2, linestyle='--', linewidth=2, label="Dashed Line")
plt.plot(x, y3, linestyle=':', linewidth=2, label="Dotted Line")

plt.xlabel="X Axis"
plt.ylabel="Y Axis"
plt.title="Line Plot with Different Styles"

plt.legend()

plt.show()
```

```python
# 8. Write a Python program to plot two or more lines and set the line markers.

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [8, 18, 28, 38, 48]
y3 = [5, 15, 25, 35, 45]

plt.plot(x, y1, marker='o', label="Circle Marker", color = 'orange')
plt.plot(x, y2, marker='s', label="Square Marker", color = 'black')
plt.plot(x, y3, marker='^', label="Triangle Marker", color = 'green')

plt.xlabel="X Axis"
plt.ylabel="Y Axis"
plt.title="Line Plot with Different Markers"

plt.legend()

plt.show()
```

```python
# 9. Write a Python program to display the current axis limits values and set new axis values.

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y)

print("Current X-axis limits:", plt.xlim())
print("Current Y-axis limits:", plt.ylim())

plt.xlim(0, 6)
plt.ylim(0, 60)

plt.xlabel="X Axis"
plt.ylabel="Y Axis"
plt.title="Displaying and Setting Axis Limits"

plt.show()
```
