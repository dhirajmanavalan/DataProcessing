```python
# 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,label = 'Problem-1')
plt.xlabel = 'x'
plt.ylabel = 'y'
plt.title = 'draw a line'
plt.show()
```

```python
# 2. Write a Python program to draw a line using given axis values with suitable label in the x axis, y axis and a title

x = [10,20,30,40,50]
y=[60,70,80,90,100]
plt.plot(x,y, label = 'Line Graph')
plt.xlabel = 'x axis'
plt.ylabel = 'y axis'
plt.title = 'Line plot'
plt.legend()
plt.show()
```

```python
# 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
# Test Data:
# test.txt
# 1 2
# 2 4
# 3 1

x = [1,2,3]
y=[2,4,1]

plt.plot(x,y, label = 'Line Graph')
plt.xlabel = 'x'
plt.ylabel = 'y'
plt.title = 'Graph'
plt.legend()
plt.show()
```

```python
x = []
y = []

with open(r"test.txt", "r") as file:
    for line in file:
        a, b = line.split()
        x.append(int(a))
        y.append(int(b))

plt.plot(x, y)
plt.xlabel = 'x'
plt.ylabel = 'y'
plt.title = 'line graphh'
plt.show()
```

```python
# 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [5, 6, 7, 8, 9]

plt.plot(x, y1, label="Line 1", color = 'orange')
plt.plot(x, y2, label="Line 2" , color = 'black')
plt.plot(x, y3, label="Line 3", color = 'red')

plt.xlabel="X Axis"
plt.ylabel="Y Axis"
plt.title="Multiple Line Plot"

plt.legend()

plt.show()
```
