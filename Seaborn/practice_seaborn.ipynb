```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
sns.set_style('ticks')
sns.set_context("notebook")
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='island', palette="Set1")
sns.despine()
```

```python
sns.scatterplot(data=penguins,x='species',y='body_mass_g',hue='island',style='sex',alpha=0.5)
```

```python
sns.stripplot(data=penguins, x='species',y='body_mass_g',hue='island',dodge = True,jitter=True)
```

```python
# swarm plot
sns.swarmplot(data=penguins,x='species',y='body_mass_g',hue='island')
```

```python
# histogram
sns.histplot(data=penguins,x='body_mass_g',hue='sex',multiple='stack')
```

```python
# regression plot
sns.regplot(data=penguins,x='body_mass_g',y='flipper_length_mm',color='red')
```

```python
# line plot
sns.set_style('darkgrid')
sns.lineplot(data=penguins,x='body_mass_g',y='flipper_length_mm',hue='island',style='sex')
```

```python
# join plots
sns.jointplot(data=penguins,x='body_mass_g',y='flipper_length_mm')
```

```python
sns.jointplot(data=penguins,x='body_mass_g',y='flipper_length_mm',hue = 'sex',kind='kde')
```

```python
# Bar plot
sns.barplot(data=penguins,x='species',y='body_mass_g',hue='sex')
```
